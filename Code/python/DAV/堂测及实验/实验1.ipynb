```python
import numpy as np
np.set_printoptions(
    suppress=True,      # 禁用科學記號（如 1e+05）
    precision=2,        # 小數點後 2 位
    linewidth=150,      # 每行最大字元數（預設 75，太小會換行）
    threshold=np.inf    # 顯示所有元素（避免中間用 ... 省略）
)
hcz1 = np.linspace(0, 0.95, 20).reshape(4, 5)
print(hcz1)
```

```python
hcz2 = np.random.rand(1, 5)
print(hcz2)
```

```python
print(hcz1+hcz2,"\n")
print(hcz1-hcz2,"\n")
print(hcz1*hcz2,"\n")
print(hcz1/hcz2,"\n")
```

```python
tmp = [[2012, 538580.0, 49084.6, 244639.1, 244856.2],
[2013, 592963.2, 53028.1, 261951.6, 277983.5],
[2014, 643563.1, 55626.3, 277282.8, 310654.0],
[2015, 688858.2, 57774.6, 281338.9, 349744.7],
[2016, 746395.1, 60139.2, 295427.8, 390828.1],
[2017, 832035.9, 62099.5, 331580.5, 438355.9],
[2018, 919281.1, 64745.2, 364835.2, 489700.8],
[2019, 986515.2, 70473.6, 380670.6, 535371.0],
[2020, 1013567.0, 78030.9, 383562.4, 551973.7],
[2021, 1143669.7, 83085.5, 450904.5, 609679.7]]
hcz3 = np.array(tmp)
print (hcz3)
```

```python
year = hcz3[:, 0:1]   # 年份
hcz4 = hcz3[:, 2:3]   # 第一
hcz5 = hcz3[:, 3:4]   # 第二
hcz6 = hcz3[:, 4:5]   # 第三
sumG = hcz3[:, 1:2]   # 總值
gdp1 = np.round((hcz4 / sumG), 2)
gdp2 = np.round((hcz5 / sumG), 2)
gdp3 = np.round((hcz6 / sumG), 2)
print("佔比分別爲(12-21年)\n第一產業\n",gdp1,"\n第二產業\n",gdp2,"\n第三產業\n",gdp3)
```

```python
hcz7 = np.hstack((year, sumG, hcz4, gdp1, hcz5, gdp2, hcz6, gdp3))
print (hcz7)
```
